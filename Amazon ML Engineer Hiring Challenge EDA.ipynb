# Importing important Libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# Importing Data:

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

# Data Preprocessing:

```python
# Let's make a copy of both test and train data:
tr = train.copy()
te = test.copy()
```

## 1. For Train Data:

```python
# let's take a look at the missing data
train.isna().sum()
```

```python
tr.describe()
```

```python
# Filling missing data for numerical feature
tr['customer_product_search_score'].fillna(tr['customer_product_search_score'].mean(),inplace=True)
tr['customer_stay_score'].fillna(tr['customer_stay_score'].mean(),inplace=True)
tr['customer_product_variation_score'].fillna(tr['customer_product_variation_score'].median(),inplace=True)
tr['customer_order_score'].fillna(tr['customer_order_score'].median(),inplace=True)
```

```python
tr.isna().sum()
```

```python
# making a new copy 
tr1 = tr.copy()
```

```python
# Following feature I am filling after careful data analysis beforehand
for i in range(len(tr1['customer_visit_score'])):
    if tr1['customer_active_segment'][i] is np.nan:
        if tr1['customer_visit_score'][i] <= 10:
            tr1['customer_active_segment'][i] =  'D'
        elif tr1['customer_visit_score'][i] > 10 and tr1['customer_visit_score'][i] <= 30:
            tr1['customer_active_segment'][i] = 'B'
        elif tr1['customer_visit_score'][i] >= 30:
            tr1['customer_active_segment'][i] = 'AA'
    else:
        pass
```

```python
tr1.isna().sum()
```

```python
for i in range(len(tr1['customer_order_score'])):
    if tr1['X1'][i] is np.nan:
        if tr1['customer_order_score'][i] <= 3 and tr1['customer_order_score'][i] >2:
            tr1['X1'][i] =  'BA'
        if tr1['customer_order_score'][i] <= 2:
            tr1['X1'][i] =  'F'
        elif tr1['customer_order_score'][i] <= 4 and tr1['customer_visit_score'][i] > 3:
            tr1['X1'][i] = 'A'
        elif tr1['customer_order_score'][i] > 4 and tr1['customer_visit_score'][i] <= 6:
            tr1['X1'][i] = 'AA'
        elif tr1['customer_order_score'][i] > 6:
            tr1['X1'][i] = 'E'
    else:
        pass
```

```python
tr1.isna().sum()
```

```python
tr1['X1'].fillna('A',inplace=True)
```

```python
tr1.isna().sum()
```

### So, now our train data is clean.

## 1. Correlation between the features:

```python
cor = tr1.corr()
a = (12, 10)

fig, ax = plt.subplots(figsize=a)
sns.heatmap(cor).set_title("Correlation Matrix",fontsize=50)
```

### What we got from correlation matrix:
#### 1. *Positvely correlated:-*

* **customer_ctr_score is positively correlated to customer_stay_score:** Customer who stays more in the website are more likely to click more searched link. 


* **customer_frequency_score is positively correlated to customer_product_variation_score:** customer whose frequency of visit in one day is more are more likely to searche variety of products.


* **customer_affinity_score is positively correlated to customer_order_score:** customers who like the website loves to order more from the website. Trust is there.


* **ustomer_ctr_score is positively correlated to customer_product_variation_score:** customer who like to seacher variety of the product do more search by cliking more links.


#### 2. *Negetively correlated:*

* **customer_order_score is negatively correlated to customer_product_variation_score:** customer who searches variety product usually do not buy them. Customer are just checking the products.


* **customer_visit_score is negatively correlated to customer_stay_score:** Customer who visit the sites regularly do not stay for long. The reason behide this can be the just come for a little to see some deals in the product and then left. Since they visit regularly they do not need to stay more time to know the deals.
